```python
df = pd.DataFrame()
for i in range(1,8):
    csv_path = rf"C:\Users\Abdallah Mohamed\Desktop\jobs\page{str(i)}.csv"
    csv_reader = pd.read_csv(csv_path)
    df = df._append(csv_reader)
```

```python
DATAFRAME_LENGTH = len(df)
df_index = pd.Index(range(DATAFRAME_LENGTH))
df.set_index(df_index, inplace=True)
```

```python
df.tail()
```

```python
dfWithoutYears = df.loc[:, 'Years of experience'].str.replace('Years',' Years')
minYearsCol = dfWithoutYears.str.split().str[0].astype("int64")
maxYearsCol = dfWithoutYears.str.split().str[2].astype("int64")
df['Min Years of Experience'] = minYearsCol
df['Max Years of Experience'] = maxYearsCol
```

```python
df
```

```python
mask = (df.loc[:,'Salaries'].str.contains(r'\d', regex=True, na =False))
df.loc[:, 'Salaries'] = df.loc[:,'Salaries'].where(mask,np.nan) 
```

```python
DATAFRAME_LENGTH = len(df)
df_index = pd.Index(range(DATAFRAME_LENGTH))
df.set_index(df_index, inplace=True)
```

```python
SalariesCol = df.loc[:, 'Salaries'].str.replace('USD ','')
minSalaryCol = SalariesCol.str.split().str[0].astype("float64")
maxSalaryCol = SalariesCol.str.split().str[2].astype("float64")
df['Min Salary'] = minSalaryCol
df['Max Salary'] = maxSalaryCol
```

```python
df
```

```python
SkillsCol = df.loc[:, 'Skills']
SkillsCol = SkillsCol.str.replace("[",'')
SkillsCol = SkillsCol.str.replace("]",'')
SkillsCol = SkillsCol.str.replace("'",'')
df.loc[:, 'Skills'] = SkillsCol
```

```python
df
```

```python
LocationsCol = df.loc[:, 'Locations']
LocationsCol = LocationsCol.str.replace("[",'')
LocationsCol = LocationsCol.str.replace("]",'')
LocationsCol = LocationsCol.str.replace("'",'')
df.loc[:, 'Locations'] = LocationsCol
```

```python
df
```

```python
NewOrder = ["Company Names",
            "Locations",
            "Years of experience",
            "Min Years of Experience",
            "Max Years of Experience",
            "Skills",
            "Salaries",
            "Min Salary",
            "Max Salary",
            "Links"]
```

```python
df = df.reindex(columns=NewOrder)
```
